```python
import pandas as pd 
import numpy as np
```

# TASK 1

```python
df=pd.read_csv('Car_sales.csv')
df
```

```python
df.info()
```

```python
df.drop(columns='__year_resale_value',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
df
```

```python
df.index=range(152)
```

```python
df
```

```python
df['Latest_Launch']=df['Latest_Launch'].astype('datetime64[ns]')
```

```python
df['Latest_Launch'].dtypes
```

```python
numerical=[]
for x in df.columns:
    if df[x].dtypes != 'O' and df[x].dtypes != '<M8[ns]' :
        numerical.append(x)
    else:
        continue
```

```python
df1=df[numerical]
df1
```

```python
df1.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
df2=StandardScaler().fit_transform(df1)
df2
```

```python
df3=pd.DataFrame(df2,columns=df1.columns)
df3
```

# TASK 2

```python
import matplotlib.pyplot as plt

for x in range(len(df3.columns)-1):
    for y in range(x+1,len(df3.columns)):
        plt.scatter(x=df3.iloc[:,x],y=df3.iloc[:,y])
        plt.xlabel(df3.head(0).iloc[:,x].name)
        plt.ylabel(df3.head(0).iloc[:,y].name)
        plt.show()
```

```python
from sklearn.cluster import KMeans

for x in range(len(df3.columns)-1):
    for y in range(x+1,len(df3.columns)):
        df4  = df3[[df3.head(0).iloc[:,x].name,df3.head(0).iloc[:,y].name]]
        kmeans = KMeans(n_clusters=3,n_init=3).fit(df4)
        centroids = kmeans.cluster_centers_
        print('\n',df3.head(0).iloc[:,x].name,'VS',df3.head(0).iloc[:,y].name)
        print('\n')
        print(centroids)
        plt.scatter(df4[df3.head(0).iloc[:,x].name], df4[df3.head(0).iloc[:,y].name], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
        plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
        plt.xlabel(df3.head(0).iloc[:,x].name)
        plt.ylabel(df3.head(0).iloc[:,y].name)
        plt.show()
```

# TASK 3

```python
df4  = df3[['Width','Length']]
kmeans = KMeans(n_clusters=3,n_init=3).fit(df4)
centroids = kmeans.cluster_centers_
print(centroids)
print('Attributes used for cluster model: Width,Length')
plt.scatter(df4['Width'], df4['Length'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
```

# TASK 4

```python
df4  = df3[['Sales_in_thousands','Price_in_thousands']]
kmeans = KMeans(n_clusters=3,n_init=3).fit(df4)
centroids = kmeans.cluster_centers_
print(centroids)
print('Attributes used for cluster model: Sales_in_thousands , Price_in_thousands')
plt.scatter(df4['Sales_in_thousands'], df4['Price_in_thousands'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.xlabel('Sales_in_thousands')
plt.ylabel('Price_in_thousands')
plt.show()
```
