```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('Car_sales.csv')
df
```

```python
df.drop(columns='__year_resale_value',axis=1,inplace=True)
df.dropna(inplace=True)
df
```

```python
df.index=range(len(df))
df
```

```python
df['Latest_Launch']=df['Latest_Launch'].astype('datetime64[ns]')
df['Latest_Launch'].dtypes
```

```python
numerical=[]
for x in df.columns:
    if df[x].dtypes != 'O' and df[x].dtypes != '<M8[ns]' :
        numerical.append(x)
    else:
        continue
```

```python
df1=df[numerical]
df1
```

```python
from sklearn.preprocessing import StandardScaler
df2=StandardScaler().fit_transform(df1)
df2
```

```python
df3=pd.DataFrame(df2,columns=df1.columns)
df3
```

# TASK 1

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm

distorsions = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    clusterer=kmeans.fit(df3)
    distorsions.append(kmeans.inertia_)
    cluster_labels = clusterer.predict(df3)
    silhouette_avg = silhouette_score(df3, cluster_labels)
    print("For k =",k,"The average silhouette_score is :", silhouette_avg)
fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 11), distorsions)
plt.grid(True)
plt.title('Elbow curve')
plt.show()
```

```python
from sklearn.cluster import KMeans

for x in range(len(df3.columns)-1):
    for y in range(x+1,len(df3.columns)):
        df4  = df3[[df3.head(0).iloc[:,x].name,df3.head(0).iloc[:,y].name]]
        kmeans = KMeans(n_clusters=3,n_init=3).fit(df4)
        centroids = kmeans.cluster_centers_
        print('\n',df3.head(0).iloc[:,x].name,'VS',df3.head(0).iloc[:,y].name)
        print('\n')
        print(centroids)
        plt.scatter(df4[df3.head(0).iloc[:,x].name], df4[df3.head(0).iloc[:,y].name], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
        plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
        plt.xlabel(df3.head(0).iloc[:,x].name)
        plt.ylabel(df3.head(0).iloc[:,y].name)
        plt.show()
```
