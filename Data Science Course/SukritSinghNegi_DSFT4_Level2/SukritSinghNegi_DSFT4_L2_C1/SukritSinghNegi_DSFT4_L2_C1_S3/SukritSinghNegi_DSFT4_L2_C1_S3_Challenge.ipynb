# Web Scraping With Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://store.steampowered.com/tags/en/Action/'
```

```python
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,'lxml')
soup
```

```python
soup.find_all('a')
```

```python
sub_url=soup.find('a', class_="submenuitem").get('href')
sub_url
```

```python
resp=requests.get(sub_url)
resp
```

```python
data=BeautifulSoup(resp.content,'lxml')
data.find_all('a')
```

```python
a=data.find_all('a',class_='gutter_item')
a
```

```python
link=[]
for b in a:
    link.append([b.get('href')])
link
```

```python
name=[]
for b in a:
    name.append(b.text.replace('\r\n\t\t\t\t\t\t\t',''))
name
```

```python
name1=[]
for i in name:
    name1.append([i.replace('\t\t\t\t\t\t','')])
name1
```

```python
table=[]
for a,n in zip(link,name1):
    table.append(n+a)
table
```

```python
table=pd.DataFrame(table,columns=['name','link'])
table
```

```python
selling=top_selling.find_all('a' ,class_="search_result_row ds_collapse_flag")
selling
```

```python

data=[]
for i in selling:
#     print(i.text.split('\n\n'))
    data.append(i.text.split('\n'))
```

```python
for a in range(5):
    for i in data:
        for j in i:
            if j==''or j==' 'or j=='\r' or j=='VR Supported ':
                i.remove(j)
data
```

```python
space=['                   ','     ','  ']
for i in range(len(data)):
    for j in range(len(data[i])):
        for k in space:
            data[i][j]=data[i][j].replace(k,'')
```
