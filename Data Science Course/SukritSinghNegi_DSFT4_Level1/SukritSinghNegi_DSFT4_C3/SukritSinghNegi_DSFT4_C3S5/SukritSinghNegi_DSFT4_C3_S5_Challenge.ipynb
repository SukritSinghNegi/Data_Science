```python
def a(radian):#Radians to Degrees
    pi=22/7
    degree=(radian*180)/pi
    return degree

def b(degree):#Degrees to Radians
    pi=22/7
    radian=(degree*pi)/180
    return radian

def c(num):#Number to its Factorial
    fact=1
    while(num>0):
        fact=fact*num
        num=num-1
    return fact

def d(num):#Number to its Square Root
    root=num**0.5
    return root

def e(mega):#Megabyte to Nibble
    nibble=mega*2000000
    return nibble

def f(frac1,frac2):#Fraction to Decimal
    dec=frac1/frac2
    return dec

def g(km_hr):# Kilometre/Hour to Meter/Second
    m_s=(km_hr*18)/5
    return m_s

def h(m_s):# Meter/Second to Kilometre/Hour
    km_hr=(m_s*5)/18
    return km_hr

def i(Int):# integer to float 
    Float=float(Int)
    return Float

def j(Float):# float to integer
    Int=int(Float)
    return Int

def k(dec):# Decimal to Octal
    return oct(dec)

def l(Oct):# Octal to decimal
    return int(Oct)

def m(dec):# Decimal to Hexadecimal
    return hex(dec)

def n(Hex):# hexadecimal to decimal
    return int(Hex)

def o(dec):# Decimal to Binary
    return bin(dec)

def p(Bin):# Binary to decimal
    return int(Bin)
    
def q(char):# Character to ASCII
    return ord(char)

def r(ASCII):# ASCII to Character
    return chr(ASCII)

def s(cel): # Celsius to Fahrenheit
    fer=(cel*1.8)+32
    return fer

def t(fer): # Fahrenheit to Celsius
    cel=(fer-32)/1.8
    return cel

# Menu
print('Conversion Menu:')
print(' a)Radians to Degrees\n b)Degrees to Radians\n c)Number to its Factorial\n d)Number to its Square Root\n e)Megabyte to Nibble\n f)Fraction to Decimal\n g)Kilometre/Hour to Meter/Second\n h)Meter/Second to Kilometre/Hour\n i)Integer to Float\n j)Float to Integer\n k)Decimal to Octal\n l)Octal to Decimal\n m)Decimal to Hexadecimal\n n)Hexadecimal to Decimal\n o)Decimal to Binary Conversion\n p)Binary to Decimal\n q)Character to ASCII\n r)ASCII to Character\n s)Celsius to Fahrenheit\n t)Fahrenheit to Celsius\n\n')

choice_1=str(input('Enter the Alphabet: '))
choice=choice_1.lower()
print('Your choice',choice)

if choice=='a':
    radian=float(input('Enter the radian: '))
    print('degree :',a(radian))
    
elif choice=='b':
    degree=float(input('Enter the degree: '))
    print('radian: ',b(degree))
    
elif choice=='c':
    num=int(input('Enter the number: '))
    print('factorial: ',c(num))
    
elif choice=='d':
    num=int(input('Enter the number: '))
    print('square root: ',d(num))
    
elif choice=='e':
    mega=int(input('Enter the megabyte: '))
    print('nibble: ',e(mega))
    
elif choice=='f':
    frac1=int(input('Enter the number: '))
    frac2=int(input('Enter the number: '))
    print(frac1,'/',frac2)
    print('decimal: ',f(frac1,frac2))
    
elif choice=='g':
    km_hr=float(input('Enter the km/hr: '))
    print('m/sec: ',g(km_hr))
    
elif choice=='h':
    m_s=float(input('Enter the m/sec: '))
    print('km/hr: ',h(m_s))
    
elif choice=='i':
    Int=int(input('Enter the integer: '))
    print('Float: ',i(Int))
    
elif choice=='j':
    Float=float(input('Enter the float: '))
    print('Int: ',j(Float))
    
elif choice=='k':
    dec=int(input('Enter the decimal: '))
    print('octal: ',k(dec))   
    
elif choice=='l':
    Oct=int(input('Enter the octal: '),8)
    print('decimal: ',l(Oct)) 
    
elif choice=='m':
    dec=int(input('Enter the decimal: '))
    print('Hexadecimal: ',m(dec)) 
    
elif choice=='n':
    Hex=int(input('Enter the hexadecimal: '),16)
    print('decimal: ',n(Hex)) 
    
elif choice=='o':
    dec=int(input('Enter the decimal: '))
    print('binary: ',o(dec))
    
elif choice=='p':
    Bin=int(input('Enter the binay: '),2)
    print('decimal: ',p(Bin))
    
elif choice=='q':
    char=str(input('Enter the character: '))
    print('ASCII: ',q(char))
    
elif choice=='r':
    ASCII=int(input('Enter the ASCII: '))
    print('char: ',r(ASCII))
    
elif choice=='s':
    cel=float(input('Enter the celsius: '))
    print('Fahrenheit: ',s(cel))
    
elif choice=='t':
    fer=float(input('Enter the Fahrenheit: '))
    print('celsius: ',t(fer))
else:
    print('Wrong Choice')
```
