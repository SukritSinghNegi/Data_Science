# Creating and Manipulating NumPy Arrays

```python
import numpy as np
```

```python
# Task 1

hr=np.arange(1,501)
finance=np.arange(501,1001)
it=np.arange(1001,1501)
sales=np.arange(1501,2001)
print('Department HR\n',hr,'\n\n')
print('Department Finance\n',finance,'\n\n')
print('Department IT\n',it,'\n\n')
print('Department Sales\n',sales,'\n\n')
```

```python
# Task 2
#salary for HR
print('Annual Salary of HR Department Employees')
print('number of employees in HR department: ',hr.size)
sal_hr=np.random.randint(25000,50000,500)
print('\nSalary for HR department employees:\n ',sal_hr)

#salary for Finance
print('\nAnnual Salary of Finance Department Employees')
print('number of employees in Finance department: ',finance.size)
sal_finance=sal_hr*1.25
print('\nSalary for Finance department employees:\n ',sal_finance)

#salary for  Sales
print('\nAnnual Salary of Sales Department Employees')
print('number of employees in Sales department: ',sales.size)
sal_sales=sal_finance*0.5
print('\nSalary for Sales department employees:\n ',sal_sales)

#salary for IT
print('\nAnnual Salary of IT Department Employees')
print('number of employees in IT department: ',it.size)
sal_it=sal_hr+5000
print('\nSalary for Sales department employees:\n ',sal_it)

```

```python
# Task 3
hrdep_comp=sal_hr.sum()
print('HR Department total expences: ',hrdep_comp)
findep_comp=sal_finance.sum()
print('Finance Department total expences: ',findep_comp)
saldep_comp=sal_sales.sum()
print('Sales Department total expences: ',saldep_comp)
itdep_comp=sal_it.sum()
print('IT Department total expences: ',itdep_comp)

total_comp=np.array([hrdep_comp,findep_comp,saldep_comp,itdep_comp])
maximum=total_comp.max()
print('The department costing more by amount:',total_comp.max())

dic_comp={'hr':hrdep_comp,'Finance':findep_comp,'Sales':saldep_comp,'IT':itdep_comp}
for x in dic_comp:
    if dic_comp[x]==maximum:
        print('Department costing more to the company is',x)
```

```python
# Task 4
tsal_hr=np.sort(sal_hr)[-1:-11:-1]
print('Top 10 of HR Department:\n',tsal_hr)
print('The Sum of Top 10 of HR Department:',tsal_hr.sum())

tsal_fin=np.sort(sal_finance)[-1:-11:-1]
print('\nTop 10 of Finance Department:\n',tsal_fin)
print('The Sum of Top 10 of Finance Department:',tsal_fin.sum())

tsal_sales=np.sort(sal_sales)[-1:-11:-1]
print('\nTop 10 of Sales Department:\n',tsal_sales)
print('The Sum of Top 10 of Sales Department:',tsal_sales.sum())

tsal_it=np.sort(sal_it)[-1:-11:-1]
print('\nTop 10 of IT Department:\n',tsal_it)
print('The Sum of Top 10 of IT Department:',tsal_it.sum())
```

```python
# Task 5
print('The count of the employees with salary greater in HR :',np.count_nonzero(sal_hr>45000))
print('The count of the employees with salary greater in Finance :',np.count_nonzero(sal_finance>45000))
print('The count of the employees with salary greater in Sales :',np.count_nonzero(sal_sales>45000))
print('The count of the employees with salary greater in IT :',np.count_nonzero(sal_it>45000))
abc_sal=np.append(sal_hr,sal_finance)
abc_sal=np.append(abc_sal,sal_it)
abc_sal=np.append(abc_sal,sal_sales)
print(abc_sal)
print('\nThe count of the employees with salary greater :',np.count_nonzero(abc_sal>45000))
abc_sal=abc_sal*1.1
print('\nThe salary increamented by 10% \n',abc_sal)
```

```python
# Task 6
index=np.where(sal_sales<20000)# index 
inc_salsales=sal_sales[index]
inc_salsales=inc_salsales*1.2
x_salsales=np.delete(sal_sales,(index))
inc_salsales=np.append(x_salsales,inc_salsales)
print('The new sale salary\n',inc_salsales)
print('The size of the salary :',inc_salsales.size)
print(sal_sales)
sal_sales=inc_salsales.copy()
print(sal_sales)

```

```python
# Task 7
index=np.where(sales>=1996)
sales=np.delete(sales,(index))
sales_1=np.arange(2001,2006)
sales=np.append(sales,sales_1)
print('The last ten enteries:',sales[-1:-11:-1])
```
