```python
# Task 1
a=df[df.year==2014]
a1=a.groupby(['make'])['sellingprice'].mean()
print(type(a1))
a1=a1.sort_values()
a1.head()
```

```python
# Task 2
b=df[(df.year==2010)|(df.year==2015)&(df.condition>4.5)]
b1=b.groupby(['year'])['sellingprice'].mean()
print(b1)
increase=((b1.values[1]-b1.values[0])/b1.values[0])
print('The percentage by which the avereage selling price of 2015 model are more the 2010 model ',increase*100,'%')
```

```python
# Task 3
print(df['interior'].value_counts())
# df['interior']=df['interior'].fillna('black')
df.isnull().sum()
```

```python
c=df[(df.interior=='black')|(df.interior=='gray')]
c1=c.groupby(['interior'])['mmr'].mean()
print(c1)
increase=((c1.values[0]-c1.values[1])/c1.values[1])
print('The mmr of black interior is more then gray by',increase*100,'%')
```

```python
# Task 4
d=df[(df.year>=2013)&(df.year<=2015)]
d1=pd.pivot_table(d,index=['color','make'],values=['sellingprice'],aggfunc='mean')
d2=pd.pivot_table(d,index=['color','make'],values=['sellingprice'],aggfunc='sum')
d3=d.groupby(['color'])['make'].count().sort_values(ascending=False)
d4=d.groupby(['color','make'])['make'].count().sort_values(ascending=False)
print(d4)
print(d3)
print('\n',d1['sellingprice'].sort_values(ascending=False).head())
print('\n',d2['sellingprice'].sort_values(ascending=False).head())
print('\nThe color of cars people use most often is:',d3.index[0])
```

```python
# Task 5
e=df[(df.body=='SUV')&(df.saledate=='Fri Dec 19 2014 09:00:00 GMT-0800 (PST)')|(df.saledate=='Thu Jan 15 2015 04:30:00 GMT-0800 (PST)')]
print('Total_sale:',e['sellingprice'].sum())
e1=pd.pivot_table(e,index=['saledate'],values=['sellingprice'],aggfunc='sum')
print(e1)
increase=(e1.values[1]-e1.values[0])/e1.values[0]
print('\nThe increase in the sales is by ',increase*100,'%')
```

```python
# Task 6
discountedprice=list(df.sellingprice)
df.insert(15,'discountedprice',discountedprice,True)
```

```python
f1=df[(df.make=='Ford')|(df.make=='Chevrolet')]
f1['discountedprice']=f1['discountedprice']*0.9
df[(df.make=='Ford')|(df.make=='Chevrolet')]=f1
f2=df[(df.make!='Ford')|(df.make!='Chevrolet')]
f2['discountedprice']=f2['discountedprice']*0.8
df[(df.make!='Ford')|(df.make!='Chevrolet')]=f2
df
```
