```python
# Task 1
df.duplicated(keep=False)
```

```python
df1=df.loc[:,df.columns!='student_id']
df_dup=df1[df1.duplicated(keep='first')]
print(df_dup)
print('Shape before dealing with duplicates:',df1.shape)
df_ddup=df1.drop_duplicates()
print('\n')
print(df_ddup)
print('Shape after dealing with duplicates:',df_ddup.shape)
```

```python
# Task 2
df_dup1=df1[df1.duplicated(['school_type','teaching_method','pretest','posttest'],keep='first')]
print(df_dup1)
print('Shape before dealing with duplicates:',df1.shape)
df_ddup1=df1.drop_duplicates(['school_type','teaching_method','pretest','posttest'],keep='first')
print('\n')
print(df_ddup1)
print('Shape after dealing with duplicates:',df_ddup1.shape)
```

```python
# Task 3
a=df1.groupby(by='school')['pretest'].mean()
c=a.sort_values(ascending=False).head()
print('\nmaximum',c)
```

```python
a1=pd.pivot_table(df1,index=['school'],values=['pretest'],aggfunc='mean')
# print(a1)
# a1.info()
a2=a1.sort_values(by='pretest',ascending=False).head()
print('maximum :',a2)
```

```python
# Task 4
b1=pd.crosstab(df1.school_setting,df.gender,rownames=['school_setting'],colnames=['gender'])
print('Minimum female students are in',b1.index[0],':',b1.values[0][0])
print('Maximum male students are in',b1.index[2],':',b1.values[2][1])
b1
```

```python
b2=df1.groupby(by=['school_setting','gender'])['gender'].count()
print('Minimum ',b2.index[0],b2.values[0])
print('Maximum ',b2.index[-1],b2.values[-1])
```

```python
# Task 5
c1=pd.pivot_table(df1,index=['school'],values=['posttest'],aggfunc='max')
print(c1)
```

```python
sort=c1.sort_values(by='posttest',ascending=False)
sort.head()
```

```python
c2=pd.pivot_table(df1,index=['school'],values=['posttest'],aggfunc='min')
sort1=c2.sort_values(by='posttest')
sort1.head()
```
