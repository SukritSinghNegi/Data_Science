```python
# Task 1
print('Number of empty rows',df['City_Mileage_km_litre'].isnull().sum())
print('Number of empty rows',df['Make'].isnull().sum())
```

```python
a=df[(df['Make'].isnull())]
# accessing null values 
```

```python
print(df['Make'].count())
print(df['Make'].value_counts().head())
a1=df['City_Mileage_km_litre'].mean()
print('Average city mileage',a1)
```

```python
a2=df.groupby(['Make'])['City_Mileage_km_litre'].mean()
a3=a2.sort_values(ascending=False).head(10)
a3
```

```python
plt.barh(a3.index,a3.values,color='greenyellow')
plt.title('Top 10 cars in terms of fuel economy')
plt.ylabel('Cars')
plt.xlabel('Average Mileage')
plt.show()
```

Maruti Suzuki have the maximum Fuel economy

```python
# Task 3
c=df['Body_Type'].value_counts()
c1=c.sort_values(ascending=False)
print(c1)
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.pie(c1.values,labels=c1.index,autopct='%.2f%%')
plt.title('Popular Car type')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.barh(c1.index,c1.values,color='orange')
plt.title('Top 10 cars in terms of fuel economy')
plt.ylabel('Cars')
plt.xlabel('Average Mileage')
plt.show()
```
