# Predicting Events Using Discrete Probability Distribution

```python
# library
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics as st
import seaborn as sns
import scipy.stats as sts
from scipy.stats import binom
import math
```

```python
# Task 1
X=[6,7,8,9,10,11,12,13,14]
P_X=[0.03,0.08,0.15,0.2,0.19,0.16,0.10,0.07,0.02]
```

```python
# Task 1 a)
test=zip(X,P_X)
sum_a=0
for i in test:
    if i[0]!=11:
#         print(i[1])
        sum_a=sum_a+i[1]
    else:
        break
print('Probability of number of lost not exceed 10 day',sum_a)
```

```python
# Task 1 b)
test=zip(X,P_X)
sum_a=0
for i in test:
#     print(i[0])
    if i[0]>=8 and i[0]<=12:
#         print(i[1])
        sum_a=sum_a+i[1]
    else:
        continue
print('Probability of number of lost 8 to 12 day',sum_a)
```

```python
# Task 1 c)
test=zip(X,P_X)
sum_a=0
for i in test:
    sum_a=sum_a+i[1]
print('Probability of number of no lost',1-sum_a)
```

```python
# task 1 d)
test=zip(X,P_X)
mean=0
for i in test:
    mean=mean+(i[0]*i[1])
print('Mean :',round(mean,2))
```

```python
test=zip(X,P_X)
var=0
for x in test:
    var=var+(((x[0]-mean)**2)*x[1])
    
sd=math.sqrt(var)
print('Standard Deviation :',round(sd,2))
```

```python
# Task 1 e)
sns.barplot(X,P_X)
plt.title('Probability distribution')
plt.xlabel('Days')
plt.ylabel('Probability crew unable to work')
plt.show()
```

```python
# Task 2

# a)
p=0.1 
q=1-p
n=20
k=np.arange(n+1)
a=0
binomial=binom.pmf(k,n,p)
for i in range(4):
    a=a+binomial[i]
print('a)The Probability Fewer than four :',round(a,2))
cumbinomial=list(binom.cdf(k,n,p))
print('  The Probability Fewer than four :',round(cumbinomial[3],2))

#b)
print('\n')
print('b)Mean :',n*p)
print('  Standard Deviation :',n*p*(1-p))

# c)
print('\n')
sns.regplot(k,binomial,binomial,order=12, ci=None)
plt.title('Orea Purchaser Probability')
plt.xlabel('Purchaser')
plt.ylabel('Binomial')
plt.show()
plt.plot(k,binomial,'s-')
plt.title('Orea Purchaser Probability')
plt.xlabel('Purchaser')
plt.ylabel('Probability')
plt.show()
```

```python
# Task 3
p=0.2
q=1-q
n=15
k=np.arange(n+1)
binomial=binom.pmf(k,n,p)

# a)
a=round(binomial[5],3)
print('a)Probability of exactly five overseas outsourcing companies',a)
plt.plot(5,binomial[5],'s-')
plt.title('exactly five overseas outsourcing companies')
plt.xlabel('number of outsourcing companies')
plt.ylabel('Probability')
plt.show()
print('\n')

# b)
b=[]
b2=[]
sum_=0
for i in range(len(binomial)):
    if i>9:
#         print(binomial[i])
        sum_=sum_+binomial[i]
        b.append(binomial[i])
        b2.append(i)
    else:
        continue
b1=round(sum_,5)
print('b)Probability of more than nine overseas outsourcing companies',b1)
plt.plot(b2,b,'s-')
plt.title('more than nine outsourcing companies')
plt.xlabel('number of outsourcing companies')
plt.ylabel('Probability')
plt.show()
print('\n')

# c)
c=round(binomial[0],3)
print('c)Probability of none overseas outsourcing companies',c)
plt.plot(0,binomial[0],'s-')
plt.title('none overseas outsourcing companies')
plt.xlabel('number of outsourcing companies')
plt.ylabel('Probability')
plt.show()
print('\n')

# d)
d=[]
d1=[]
sum_a=0
for i in range(len(binomial)):
    if i>=4 and i<=7:
        sum_a=sum_a+binomial[i]
        d.append(binomial[i])
        d1.append(i)
    else:
        continue
d2=round(sum_a,3)
print('d)Probability of four to seven inclusive overseas outsourcing companies',d2)
plt.plot(d1,d,'s-')
plt.title('four to seven inclusive overseas outsourcing companies')
plt.xlabel('number of outsourcing companies')
plt.ylabel('Probability')
plt.show()

# e)
print('\n')
sns.regplot(k,binomial,binomial,order=12,ci=None)
plt.title('Bonimial Distribution')
plt.xlabel('number of outsourcing companies')
plt.ylabel('Probability')
plt.show()
plt.plot(k,binomial,'s-')
plt.title('Bonimial Distribution')
plt.xlabel('number of outsourcing companies')
plt.ylabel('Probability')
plt.show()
print('a) single dot is ploted on the graph to the corresponding (x,y)coordinate (',5,',',round(binomial[5],3),')')
print('''b) declining curve is ploted which means the probalility decreses with the increase in number of 
   outsourcing company after the peak''')
print('c) single dot is ploted on the graph to the corresponding (x,y)coordinate (',0,',',round(binomial[0],3),')')
print('''d) declining curve is ploted which means the probalility decreses with the increase in number of 
   outsourcing company after the peak''')
```

```python
# Task 4
dis={'week1':[3,6,4,6,2,3,1,5,1,0,3,3],'week2':[1,2,4,0,2,6,5,4,2,5,3,4],'week3':[5,3,5,3,5,4,7,3,4,8,3,1]}
df=pd.DataFrame(dis)
df
```

```python
q=df.mean().mean()
n=np.arange(21)
poisson=sts.poisson.pmf(n,q)

# a)
print('a)No Customer arrive:',round(poisson[0],3))
plt.plot(0,poisson[0],'s-',label='3.5')
plt.title('No Customer arrive')
plt.xlabel('Customer arrive')
plt.ylabel('Probability')
plt.show()
print('\n')
# b)
b=[]
b1=[]
sum_b=0
for i in range(len(poisson)):
    if i>=6:
        b.append(poisson[i])
        b1.append(i)
        sum_b=sum_b+poisson[i]
    else:
        continue
b2=round(sum_b,3)
print('b)More then six Customer arrive:',b2)
plt.plot(b1,b,'s-',label='3.5')
plt.title('More then six Customer arrive')
plt.xlabel('Customer arrive')
plt.ylabel('Probability')
plt.show()
print('\n')
# c)
q1=q*2
poisson1=sts.poisson.pmf(n,q1)
c=[]
c1=[]
sum_c=0
for i in range(len(poisson1)):
    if i!=4:
        c.append(poisson1[i])
        c1.append(i)
        sum_c=sum_c+poisson1[i]
    else:
        break
c2=round(sum_c,3)
print('c)less then four Customer arrive:',c2)
plt.plot(c1,c,'s-',label='7')
plt.title('less then four Customer arrive')
plt.xlabel('Customer arrive')
plt.ylabel('Probability')
plt.show()
print('\n')
#d
d=[]
d1=[]
sum_d=0
for i in range(len(poisson1)):
    if i>=3 and i<=6:
#         print(poisson1[i])
        sum_d=sum_d+poisson1[i]
        d.append(poisson1[i])
        d1.append(i)
    else:
        continue
d2=round(sum_d,3) 
print('d)between three and six Customer arrive:',d2)
plt.plot(d1,d,'s-',label='7')
plt.title('between three and six Customer arrive')
plt.xlabel('Customer arrive')
plt.ylabel('Probability')
plt.show()
print('\n')
# e)
q2=q*3
poisson2=sts.poisson.pmf(n,q2)
print('e)Exactly eight Customer arrive:',round(poisson[8],3))
plt.plot(8,poisson[8],'s-',label='10.5')
plt.title('Exactly eight Customer arrive')
plt.xlabel('Customer arrive')
plt.ylabel('Probability')
plt.show()
print('\n')
# f)
plt.plot(n,poisson,'s-',label='3.5')
plt.plot(n,poisson1,'s-',label='7')
plt.plot(n,poisson2,'s-',label='10.5')
plt.title('Possion Distribution')
plt.xlabel('Customer arrive')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
