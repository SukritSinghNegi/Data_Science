# Hypothesis Test - 1

```python
import statistics as st
import math
import pandas as pd
import numpy as np
import scipy.stats as sts
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#Task 1 a)
# H0: mu = 33.88
# Ha: mu != 33.88
print('''Task 1 a) 
    
Null hypothesis is average rent is equal to $33.88
Alternate hypothesis is average rent in not equal to $33.88
two side tail test
Formula:
H0: mean_p=33.88
Ha: mean_p!=33.88''')


mean_p=33.88
mean_s=32.18
sd_p=1.28
n_s=49
alpha=0.05 

# Task 1 b)
print('\nTask 1 b)\n')

z_critical_n=norm.ppf(alpha/2) # Calculate -z_critical value
print('- Z_critical value : ',z_critical_n)
z_critical_p=norm.ppf(1-alpha/2) #Calculate +Z_critical values
print('+ Z_critical value : ',z_critical_p)

# Task 1c)
print('\nTask 1 c)\n')
z_statistics=(mean_s-mean_p)/(sd_p/np.sqrt(n_s))# calcuting the statistics value
print('Z_statistics value : ',z_statistics)

margin_error=z_critical_p*(sd_p/np.sqrt(n_s))# Calculating the margin of error

# Calcutating the 95% of the confidence interval
lhs=(mean_s)-margin_error #gives the value left most of confidence region 
rhs=(mean_s)+margin_error #gives the value right most of confidence region
print('\nvalue from ',lhs,'to',rhs,'is the 95% confidence level\n')
print('Task 1 d)\n')
if z_critical_n < z_statistics and z_statistics < z_critical_p:# Rule -Z_critical < Z_statistics < +Z_critical
    print('CONCLUSION : Failed to reject Null hypothsis')
else :
    print('CONCLUSION : Reject null hypothesis')
    if z_critical_n > z_statistics:
        print('''z_statistics value is less then -Z_critical value''')
    else:
        print('z_statistics value is greater then Z_critical valu')

# Task 1 e)
print('\nTask 1 e)\n')
# Calculate the P_value
p_value=norm.sf(abs(z_statistics))*2
print('p-value :',p_value)
if p_value < alpha:
    print('CONCLUSION : Reject null hypothesis(p_value < alpha)')
else:
    print('CONCLUSION : Failed to reject Null hypothsis')
print('\n')       
lower=mean_p-4*sd_p
upper=mean_p+4*sd_p
plot=np.linspace(lower,upper,49)
norm_p=norm.pdf(plot,mean_p,sd_p)
plot_s=np.linspace(lower,upper,49)
norm_s=norm.pdf(plot,mean_s,sd_p)
plt.plot(plot,norm_p,label='H0')
plt.plot(plot_s,norm_s,label='Ha')
plt.legend()
plt.show()
```
