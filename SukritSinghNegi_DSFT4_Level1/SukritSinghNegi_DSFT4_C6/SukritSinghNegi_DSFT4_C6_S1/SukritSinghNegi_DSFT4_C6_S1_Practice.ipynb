# Level 1 Analysis

Separating Categorical and Numerical Variables

```python
for column in df.columns:
        print('\n',column)
        print(df[column].unique())
        print(df[column].nunique())
        print(len(df[column].unique()))
```

```python
def seperate_data_types(df):
    categorical=[]
    continuous=[]
    for column in df.columns:
        if df[column].nunique()<100:      #Return:(True or Flase) give the size of the array and check < 100 
            categorical.append(column)
        else:
            continuous.append(column)
    return categorical,continuous

categorical , continuous =seperate_data_types(df)
# print(categorical,continuous)
table=[categorical,continuous]
print(tabulate({'Categorical':categorical,"continuous":continuous},headers=['categorical','continuous']))
```

### 1. Analysis of Attrition Flag: If the account is closed (Attired Customer) else (Existing Customer)

Catagorical Data : Mode will be used for Null value treatment

```python
def info_of_cat(col): 
    print('Unique values in ',col,' are:', df[col].unique())                # unique: returns the unique values in the col
    Mode=df[col].mode()
    print('Mode of ',col,' is ',Mode[0])                                    # mode: returns the mode of the column
    print('Number of missing values in ',col,' is ',df[col].isnull().sum()) # isnull().sum() counts the number of null in dataframe
    if df[col].isnull().sum() > 0:                                          # check if null values are present
        print('\nThere are null values in the', col,' column')
```

```python
info_of_cat("Attrition_Flag")
```

```python
fig,ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].set_title("Count of Attrition Flag")                  
percentage=df["Attrition_Flag"].value_counts()              
print('Percentage')
print(percentage)
labels = list(df["Attrition_Flag"].value_counts().index)    
print(labels)
sns.countplot(x = df["Attrition_Flag"],ax=ax[0])
plt.pie(percentage,labels = labels, autopct= "%0.2f%%")     

plt.show()
```

### 8. User defined function which returns percentage using bar graphs

```python
def bar_percentage(ax, count: "number of rows in data "):
    for bar in ax.patches:                                            
        percentage = f"{round((bar.get_height() / count) *100, 2)}%"  
                                                                            
        x = bar.get_x() + bar.get_width() /2                       
        y = bar.get_height()                                          
        ax.annotate(percentage, (x, y), va = "bottom", ha = "center")                                                                 # ha: Horizontal axis:-> how you want to keep it. can be changed to center, left, right.                                                              # annotate: used to annotate over objects. here objects are the bars 
    plt.show()
```

### User defined function for calculating mean, median, mode, standard deviation, and the count of all null values

```python
print(tabulate({'Categorical':categorical,"continuous":continuous},headers=['categorical','continuous']))
```

```python
def info_of_continuous(col):
    print('Mean of ',col,'=',st.mean(df[col]))
    print('Median of ',col,'=',st.median(df[col]))
    print('Mode of ',col,'=',st.mode(df[col]))
    print('Standard Deviation of ',col,'=',st.stdev(df[col]))
    print('Count of null values of',col,'=',df[col].isnull().sum())
```
