# Task 1



Total Labour Probability P(T)= 1 =100%                                                                                 
Part-Time Labour Probability P(PT)=0.174 = 17.4%                                                                       
Full-Time Labour Probability P(FT)= P(T)- p(PT)=0.826 = 82.6%                                                          
                                                                                                                      
Female Labour Probability P(F)=0.46 = 46%                                                                              
Male Labour Probability P(M)=P(T)-P(F)= 1-0.46 = 0.54 = 54%                                                            
                                                                                                                      
Female Part-Time Labour P(PT and F)=P(FPT)=p(F) * 0.25 = 0.46 * 0.25= 0.115 = 11.5%                                                    
                                                                                                                      
If a Part-Time worker is randomly selected=P(PT|FPT)= P(FPT)/P(PT)= 0.115/0.174 = 0.661 = 66.1%

```python
import pandas as pd
import numpy as np
import statistics as st
import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task 3
df=pd.read_csv('DS1_C5_S3_Sampling_Practice_GamesSales_Data.csv')
print(df.shape)
print(type(df))
print(df.head(0))
```

```python
print(df['Year'].value_counts())
# print(df['Publisher'].value_counts())
```

```python
year=df.Year.unique()
year
```

```python
grouped_year=df[[ 'Year', 'North_America_Sales_M$', 'Europe_Sales_M$', 'Japan_Sales_M$', 'Other_Sales_M$', 'Global_Sales_M$']].groupby(['Year'])
grouped_year
```

```python
for y in year:
    year_data=grouped_year.get_group(y)
    print(y)
    for x in year_data:
        sales=pd.DataFrame()
        if x!= 'Year':
            print(x)
            sales=sales.append(year_data.loc[:,x],ignore_index=True)
#             print(sales)
            print('Mean of ',x,':',st.mean(sales.loc[0,:]))
            print('Median of ',x,':',st.median(sales.loc[0,:]))
            print('Mode of ',x,':',st.mode(sales.loc[0,:]))
            print('Standard Deviation of ',x,':',st.stdev(sales.loc[0,:]))
            print('Kurtosis of ',x,':',sts.kurtosis(sales.loc[0,:],bias=False))
            print('Skewnewss of ',x,':',sts.skew(sales.loc[0,:],bias=False))
            sns.distplot(sales,label=x)
            plt.legend()
            plt.show()           
```

```python
# Task 3.1.4

d_df=df[['Year','Genre','Global_Sales_M$']].groupby(['Year'])
year=df.Year.unique()
# print(year)
for y1 in year:
    strat_sample_df=pd.DataFrame()
    y1_data=d_df.get_group(y1)
    print('\n',y1)
    sample_data=y1_data.sample(n=30,replace=True,random_state=1)
    strat_sample_df=strat_sample_df.append(sample_data,ignore_index=True)
#     strat_sample_df.drop(['Year'],axis=1,inplace=True)
#     print(strat_sample_df)
    strat_sample_df=strat_sample_df.groupby(['Genre'])
    gen=strat_sample_df.Genre.unique()
#     print(gen.index)
    gen1=gen.index
    for g1 in gen1:
        print(g1)
        g=strat_sample_df.get_group(g1)
        print('Mean :',st.mean(g['Global_Sales_M$']))
        print('Median  :',st.median(g['Global_Sales_M$']))
        print('Mode :',st.mode(g['Global_Sales_M$']))
#         print('Standard Deviation :',st.stdev(g['Global_Sales_M$']))
        print('Kurtosis :',sts.kurtosis(g['Global_Sales_M$'],bias=False))
        print('Skewnewss :',sts.skew(g['Global_Sales_M$'],bias=False))
        print('******************')
#         strat_sample_df.drop(['Genre'],axis=1,inplace=True)
#         sns.distplot(g['Global_Sales_M$'],label=g1)
#         plt.legend()
#         plt.show()
#         print(g)
```
