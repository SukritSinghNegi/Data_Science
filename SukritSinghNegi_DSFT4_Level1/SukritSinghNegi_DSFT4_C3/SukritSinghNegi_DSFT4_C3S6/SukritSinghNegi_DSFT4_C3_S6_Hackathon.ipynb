```python
print('  ESSENTIAL ITEMS AVAILABLE')

print('S.No  ITEM     QUANTITY   COST/ITEM')

menu=[[1,'Biscuit',5,20.50],
      [2,'Cereals',10,90.00],
      [3,'Chicken',20,200.00],
      [4,'Oats',20,39.99],
      [5,'Rice',30,79.99]]
print('1     Biscuit    ',menu[0][2],'       20.50')
print('2     Cereals    ',menu[1][2],'      90.00')
print('3     Chicken    ',menu[2][2],'      200.00')
print('4     Oats       ',menu[3][2],'      39.99')
print('5     Rice       ',menu[4][2],'      79.99')


# clone menu to bill
bill=[[1,'Biscuit'],[2,'Cereals'],[3,'Chicken'],[4,'Oats'],[5,'Rice']]

print('\n\n')
# Quantity
biscuits=int(input('How many Biscuits packets are required ?'))
if biscuits > menu[0][2]:                        # menu[0][2]=biscuit
    print('OUT OF STOCK')
    print('Current Stock',menu[0][2])            # menu[0][2]=biscuit
    b=str(input('Do you want to change the quantity? (Y/N): '))
    if b.lower()== 'y':
        biscuits=0
        biscuits=int(input('How many Biscuits packets are required ?'))
        if biscuits <= menu[0][2]:               # menu[0][2]=biscuit
            amount_bis= biscuits*20.50# amount 
            bill[0].append(biscuits)             # updating Quantity in bill
            bill[0].append(amount_bis)           # updating amount in bill
        else:
            biscuits=0
            amount_bis=0
            print('OUT OF STOCK')              
    else:
        biscuits=0
        amount_bis=0
        print('not included')                
else:
    amount_bis=biscuits*20.50# amount
    bill[0].append(biscuits)            # updating Quantity in bill
    bill[0].append(amount_bis)          # updating amount in bill



cereals=int(input('How many Cereals packets are required ?'))
if cereals > menu[1][2]:
    print('OUT OF STOCK')
    print('Current Stock',menu[1][2])
    c=str(input('Do you want to change the quantity? (Y/N): '))
    if c.lower()== 'y':
        cereals=0
        cereals=int(input('How many Cereals packets are required ?'))
        if cereals <= menu[1][2]:
            amount_cer=cereals*90.00# amount
            bill[1].append(cereals)            # updating Quantity in bill
            bill[1].append(amount_cer)         # updating amount in bill
        else:
            cereals=0
            amount_cer=0
            print('OUT OF STOCK')  
    else:
        cereals=0
        amount_cer=0
        print('not included')             
else:               
    amount_cer=cereals*90.00# amount
    bill[1].append(cereals)            # updating Quantity in bill
    bill[1].append(amount_cer)         # updating amount in bill



chicken=int(input('How many Chickens packets are required ?'))
if chicken > menu[2][2]:
    print('OUT OF STOCK')
    print('Current Stock',menu[2][2])
    ch=str(input('Do you want to change the quantity? (Y/N): '))
    if ch.lower()== 'y':
        chicken=0
        chicken=int(input('How many chicken packets are required ?'))
        if chicken <= menu[2][2]:
            amount_chick=chicken*200.00# amount
            bill[2].append(chicken)               # updating Quantity in bill
            bill[2].append(amount_chick)          # updating amount in bill
        else:
            chicken=0
            amount_chick=0
            print('OUT OF STOCK')            
    else:
        chicken=0
        amount_chick=0
        print('not included')       
else:
    amount_chick=chicken*200.00# amount
    bill[2].append(chicken)               # updating Quantity in bill
    bill[2].append(amount_chick)          # updating amount in bill


oats=int(input('How many Oats packets are required ?'))
if oats > menu[3][2]:
    print('OUT OF STOCK')
    print('Current Stock',menu[3][2])
    o=str(input('Do you want to change the quantity? (Y/N): '))
    if o.lower()== 'y':
        oats=0
        oats=int(input('How many Oats packets are required ?'))
        if oats <= menu[3][2]:
            amount_oats=oats*39.99# amount
            bill[3].append(oats)             # updating Quantity in bill
            bill[3].append(amount_oats)      # updating amount in bill
        else:
            oats=0 
            amount_oats=0
            print('OUT OF STOCK')            
    else:
        oats=0
        amount_oats=0
        print('not included')               
else:
    amount_oats=oats*39.99# amount
    bill[3].append(oats)             # updating Quantity in bill
    bill[3].append(amount_oats)      # updating amount in bill


rice=int(input('How many rice packets are required ?'))
if rice > menu[4][2]:
    print('OUT OF STOCK')
    print('Current Stock',menu[4][2])
    r=str(input('Do you want to change the quantity? (Y/N): '))
    if r.lower()== 'y':
        rice=0
        rice=int(input('How many rice packets are required ?'))
        if rice <= menu[4][2]:
            amount_rice=rice*79.99# amount
            bill[4].append(rice)             # updating Quantity in bill
            bill[4].append(amount_rice)      # updating amount in bill
        else:
            oats=0
            amount_rice=0
            print('OUT OF STOCK')             
    else:
        oats=0
        amount_rice=0
        print('not included')         
else:
    amount_rice=rice*79.99# amount
    bill[4].append(rice)             # updating Quantity in bill
    bill[4].append(amount_rice)      # updating amount in bill




# print('\n\n')                                 #Test script
# print('Individual Amount')
# print('amount_bis  :',amount_bis)
# print('amount_cer  :',amount_cer)
# print('amount_chick:',amount_chick)
# print('amount_oats :',amount_oats)
# print('amount_rice :',amount_rice)

total_qty=biscuits+cereals+chicken+oats+rice# total quantity
bill=amount_bis+amount_cer+amount_chick+amount_oats+amount_rice# bill amount

if total_qty!=0:
    print('\n\n')
    print('Fill in your delivery details')
    name=input('Enter your name : ')
    address=input('Enter your address: ')
    distance=int(input('Distance form store (km): '))
    if distance<=5 and distance>=1:
        total_bill=bill+50
    elif distance<=10 and distance>=6:
        total_bill=bill+100
    elif distance>=11:
        total_bill=bill+150
    else:
        print('Shop Billing')
        total_bill=bill+0
        
    print('\n\n')


    print('               BILL')
    print('S.No  ITEM     QUANTITY   COST/ITEM')
    if biscuits != 0 and biscuits <= menu[0][2]:
        print('1     Biscuit    ',biscuits,'       ',amount_bis)
        menu[0][2]=menu[0][2]-biscuits
    
    if cereals!=0 and cereals <= menu[1][2]:
        print('2     Cereals    ',cereals,'      ',amount_cer)
        menu[1][2]=menu[1][2]-cereals

    if chicken!=0 and chicken <= menu[2][2]:
        print('3     Chicken    ',chicken,'      ',amount_chick)
        menu[2][2]=menu[2][2]-chicken

    if oats!=0 and oats <= menu[3][2]:
        print('4     Oats       ',oats,'      ',amount_oats)
        menu[3][2]=menu[3][2]-oats

    if rice!=0 and rice <= menu[4][2]:
        print('5     Rice       ',rice,'      ',amount_rice)
        menu[4][2]=menu[4][2]-rice

    print('\n\n')
    print('Total Item Cost:\t',bill)    
    print('Total Bill Amount:\t',total_bill)
    print('\n\n')
    print('Name : ',name)
    print('Address: ',address)
    print('Have a Nice day\n\n')
    
    print('  ESSENTIAL ITEMS AVAILABLE')

    print('S.No  ITEM     QUANTITY   COST/ITEM')
    print('1     Biscuit    ',menu[0][2],'       20.50')
    print('2     Cereals    ',menu[1][2],'      90.00')
    print('3     Chicken    ',menu[2][2],'      200.00')
    print('4     Oats       ',menu[3][2],'      39.99')
    print('5     Rice       ',menu[4][2],'      79.99')
    
else:
    print('\n\n')
    print('Have a nice day')
    
```
