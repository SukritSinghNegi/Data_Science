```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('restaurant_data.csv')
```

```python
df
```

```python
data=df.groupby(['type_of_food'])['postcode'].value_counts().rename('count').reset_index()
data
```

```python
fig,ax=plt.subplots(figsize=(7,15))
sns.countplot(y='type_of_food',data=df,ax=ax)
plt.show()
```

```python
data.to_csv('Task_3.csv')
```
