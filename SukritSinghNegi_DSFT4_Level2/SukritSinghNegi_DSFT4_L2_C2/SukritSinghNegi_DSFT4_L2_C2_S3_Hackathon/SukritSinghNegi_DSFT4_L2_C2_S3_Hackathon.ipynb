```python
header=[]
for h in data1[0].find_all('th'):
    header.append(h.text.replace('\n',''))
header
```

```python
titles_2015=[]
for d in data1[1:]:
    title=d.find_all('th')
    titles_2015.append([t.text.replace('\n','') for t in title])
titles_2015
```

```python
data1_2015=pd.DataFrame(titles_2015,columns=['Title'])
data1_2015=data1_2015.fillna("The Blackcoat's Daughter")
data1_2015
```
