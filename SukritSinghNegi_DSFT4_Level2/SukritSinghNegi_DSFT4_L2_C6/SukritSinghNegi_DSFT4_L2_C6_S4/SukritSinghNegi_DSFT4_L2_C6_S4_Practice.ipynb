# TASK 2

```python
from sklearn.preprocessing import MinMaxScaler
df1=df[['Lat','Lon']]
predict=MinMaxScaler().fit_transform(df1)
stand=pd.DataFrame(predict,columns=['lat','lon'])
stand
```

# TASK 3

```python
import hdbscan
cluster_size = [30,40,50,60,70,100,200,300]

for i in cluster_size:
    clusterer = hdbscan.HDBSCAN(min_cluster_size=i, gen_min_span_tree=True)
    clusterer.fit(stand)
    stand[f"{'Cluster',i}"]=clusterer.labels_
```

```python
stand
```

```python
ncluster = []
for i in cluster_size:
    x = stand[f"{'Cluster',i}"].nunique()
    ncluster.append(x)
ncluster
```

```python
plt.plot(cluster_size,ncluster)
plt.xlabel('min_cluster_size')
plt.ylabel('no. of cluster')
plt.show()
```

```python
noutliers = []
for i in cluster_size:
    x = stand[stand[f"{'Cluster',i}"] == -1][f"{'Cluster',i}"].count()
    noutliers.append(x)
noutliers
```

```python
plt.plot(cluster_size,noutliers)
plt.xlabel('min_cluster_size')
plt.ylabel('no. of outliers')
plt.show()
```

# Task 5

```python
import numpy as np
clusterer = hdbscan.HDBSCAN(min_cluster_size=70)
clusterer.fit(stand)
c=1
fig = plt.figure(figsize=(14,14))
for quant in [0.25,0.5, 0.75, 0.9]: 
    plt.subplot(2, 2, c)
    plt.title('threshold = '+ str(quant))
    threshold = pd.Series(clusterer.outlier_scores_).quantile(quant)
    outliers = np.where(clusterer.outlier_scores_ > threshold)[0]
    plt.scatter(df3.iloc[:,0],df3.iloc[:,1], s=50, linewidth=0, c='gray', alpha=0.25)
    plt.scatter(dat.iloc[outliers,0],dat.iloc[outliers,1], s=50, linewidth=0, c='red', alpha=0.5)
    c=c +1
```
