```python
df['date'] = pd.to_datetime(df.date)
df['year'] = pd. DatetimeIndex(df['date']).year
```

```python
df['building_age']=df['year']-df['yr_built']
```

```python
df['waterfront'] = df.waterfront.astype('category')
df['view'] = df.view.astype('category')
df['condition'] = df.condition.astype('category')
df['grade'] = df.grade.astype('category')
df['zipcode'] = df.zipcode.astype('category')
```

```python
df.drop(columns=['id','date','year','lat','long','yr_built','yr_renovated'],inplace=True)
```

## TASK 2

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler

df1=StandardScaler().fit_transform(df)
X = df1[:,1:]
y = df1[:,0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.1, random_state=1)

reg=LinearRegression()
reg.fit(X_train,y_train)

predict=reg.predict(X_test)
predict
reg.score(X_test,y_test)
```

## TASK 3

```python
from sklearn.ensemble import RandomForestRegressor
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import f1_score

results = list()
for x in range(10):
    model = RandomForestRegressor(max_depth=3, random_state=x)
    model.fit(X_train,y_train)
    predict=model.predict(X_test)
    results.append(model.score(X_test,y_test))
print('Accuracy: %.3f (%.3f)' % (mean(results), std(results)))
```

```python
results
```

```python
results=list()
model = RandomForestRegressor()
cv = RepeatedKFold(n_splits=10,random_state=1)
n_scores = cross_val_score(model, X, y, cv=cv,scoring='r2', n_jobs=-1)
results.append(n_scores)
print(mean(n_scores), std(n_scores))
```
