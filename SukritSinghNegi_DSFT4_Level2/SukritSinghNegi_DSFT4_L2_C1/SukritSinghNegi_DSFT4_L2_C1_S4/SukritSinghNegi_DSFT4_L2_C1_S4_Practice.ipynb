```python
cleaned=[]
for i in tweet_df['text']:
#     print(i)
    cleaned.append(re.sub('(https:\S+|\W+|r\s+)',' ',i))
```

# TASK 4

```python
earthquake=tweet_df.loc[(tweet_df['text'].str.contains('magnitude'))&(tweet_df['location'].notnull())&(tweet_df['location']!='Send me')]
earthquake['location'].value_counts()
```

```python
print(f'Number of tweet with magnitude and location:{len(earthquake)}')
earthquake
```

```python
pattern=re.compile('\d*\.\d{1}')
magnitude=[]
for i in earthquake['text']:
    if pattern.search(i):
        print(pattern.search(i).group())
        magnitude.append(pattern.search(i).group())
    else:
        continue
magnitude
```

```python
average=list(map(float,magnitude))
average
```

```python
import statistics as sts
sts.mean(average)
```
