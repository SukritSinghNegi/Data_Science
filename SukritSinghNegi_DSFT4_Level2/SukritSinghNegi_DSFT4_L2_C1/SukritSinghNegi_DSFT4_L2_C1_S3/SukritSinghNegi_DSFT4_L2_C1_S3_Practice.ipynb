# Web Scraping With Beautiful Soup

```python
import numpy as np
import requests
from bs4 import BeautifulSoup
```

## TASK 1

```python
url= 'https://en.wikipedia.org/wiki/Leonardo_DiCaprio_filmography'
```

```python
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,'lxml')
soup
```

```python
soup.prettify()
```

```python
soup.find('table')
```

```python
table=soup.find('table',{'class':'wikitable plainrowheaders sortable'})
table
```

```python
new_table1=[]
new_table2=[]
for row in table.find_all('tr')[1:]:
    columns=row.find_all('th')
    columns1=row.find_all('td')[:2]
    new_table1.append([column.text.replace("\n","") for column in columns])
    new_table2.append([column.text.replace("\n","") for column in columns1])
print(new_table1,new_table2)
```

```python
for i in range(len(new_table1)):
    if bool(new_table1[i]):
        continue
    else:
        new_table1[i]=new_table1[i-1]
new_table1
```

```python
new_table=[]
for tab1,tab2 in zip(new_table1,new_table2):
    tab=tab1+tab2
    new_table.append(tab)
new_table    
```

```python
import pandas as pd
df = pd.DataFrame(new_table, columns=['Year','Title','Role'])
df
```

```python
df.to_html('detail.html')
```

```python
Link=[]
links=soup.find_all('a')[4:]
for link in links:
    Link.append(link.get('href'))
Link
```

```python
for null in Link:
    if bool(null):
        continue
    else:
        Link.remove(null)
Link
```

```python
new_link=[]
for l in Link:
    if l[0]=='h':
        new_link.append(l)
    else:
        continue
new_link
```

```python
article=[]
for l in new_link:
#     print(l.split('-'))
    if 'leonardo' in (l.split('-')):
        article.append(l)
article
```
