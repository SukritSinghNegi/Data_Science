```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# TASK 1

```python
df=pd.read_csv("DS3_C1_S5_SuperstoreSales_Data_Hackathon.csv",parse_dates=['Order Date'])
df.info()
```

```python
df=df.set_index(df['Order Date'])
df
```

```python
data=df.sort_index(ascending=True)
data=data.resample('M').mean()
data
```

```python
ind= int(np.round(len(data)*0.7,0))
train=data[:ind]
test=data[ind:]
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.api import SimpleExpSmoothing
from statsmodels.tsa.api import ExponentialSmoothing

decomposition=seasonal_decompose(np.asarray(data['Sales']), period=12)
d_trend=decomposition.trend
d_seasonal=decomposition.seasonal
d_residual=decomposition.resid

fig,ax = plt.subplots(5,1,figsize=(20,20))

plt.subplot(412)
plt.plot(d_trend,label='Trend')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(d_seasonal,label='Seasonal')
plt.legend(loc='best')

plt.subplot(414)
plt.plot(d_residual,label='Residual')
plt.legend(loc='best')

plt.subplot(411)
plt.plot(data['Sales'],label='sales')
plt.legend(loc='best')


plt.tight_layout()
```

# TASK 2

```python
from statsmodels.tsa.stattools import adfuller,acf,pacf
def checkStationarity(data):
    pvalue = adfuller(data)[1]   
    if(pvalue>0.05):
        msg = 'p-value={}. data is not stationary'.format(pvalue)
    else:
        msg='p-value={}. data is stationary'.format(pvalue)
        
    return(msg)
```

```python
ret=checkStationarity(train['Sales']); ret
```

```python
ret=checkStationarity(test['Sales']); ret
```

```python
ret=checkStationarity(data['Sales']); ret
```

```python
diff_temp = data['Sales'] - data['Sales'].shift()
diff_temp.dropna(inplace=True)
diff_temp
```

```python
ind= int(np.round(len(diff_temp)*0.7,0))
train=diff_temp[:ind]
test=diff_temp[ind:]
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig,ax=plt.subplots(2,1,figsize=(20,10))
plot_acf(diff_temp,ax=ax[0])
plot_pacf(diff_temp,ax=ax[1])
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
sel={}
selx={}
for p in range(1,3):
    for q in range(1,3):
        m1 = ARIMA(train,order=(p,0,q)).fit()
        sel[f'({p},0,{q})']=m1.aic
        selx[f'({p},0,{q})']=m1.llf
```

```python
min_key=min(sel,key=sel.get)
min_key
```

```python
max_key=max(selx,key=selx.get)
max_key
```

# TASK 3

```python
from statsmodels.tsa.arima.model import ARIMA

def arima(train,p,d,q,test):
    m1 = ARIMA(train,order=(p,d,q)).fit()
    predictions=m1.predict(0,len(train))
    actual=[]
    predicted=[]
    for i in range(len(test)):
        actual.append(test[i])
        predicted.append(predictions[i])
        df_res = pd.DataFrame({'actual':actual, 'predicted':predicted})
        df_res['err'] = df_res.actual - df_res.predicted
    return df_res
```

```python
comp=arima(train,2,0,2,test)
```

```python
comp
```

```python
ax=plt.subplots(figsize=(15,8))
plt.plot(comp.actual,label='actual')
plt.plot(comp.predicted,label='predicted')
plt.legend()
plt.show()
```

```python
ax=plt.subplots(figsize=(15,8))
plt.plot(comp.err,label='error')
plt.legend()
plt.show()
```
